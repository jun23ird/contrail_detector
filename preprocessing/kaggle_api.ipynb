```python
import os

# Define a function to check if a file exists
def file_exists(filepath):
    return os.path.exists(filepath)


# Specify your base directory where the files should exist
base_directory = "/content/drive/MyDrive/kaggle/contrail_detector/data/"

# Add columns to your dataframes that indicate whether the file exists
# Assuming that the 'filename' column contains the file names
train_metadata['image_exists'] = train_metadata['record_id'].apply(lambda x: file_exists(base_directory +"train/image/" + str(x)+".npy"))
valid_metadata['image_exists'] = valid_metadata['record_id'].apply(lambda x: file_exists(base_directory +"valid/image/" + str(x)+".npy"))
test_metadata['image_exists'] = test_metadata['record_id'].apply(lambda x: file_exists(base_directory + "test/image/" + str(x)+".npy"))

train_metadata['mask_exists'] = train_metadata['record_id'].apply(lambda x: file_exists(base_directory +"train/mask/" + str(x)+".npy"))
valid_metadata['mask_exists'] = valid_metadata['record_id'].apply(lambda x: file_exists(base_directory +"valid/mask/" + str(x)+".npy"))
test_metadata['mask_exists'] = test_metadata['record_id'].apply(lambda x: file_exists(base_directory + "test/mask/" + str(x)+".npy"))
```

# Add 'contrail_exists' column to the metadata dataframes

```python
# Mount your Google Drive.
from google.colab import drive
drive.mount("/content/drive", force_remount=True)
import pandas as pd
train_metadata = pd.read_json("/content/drive/MyDrive/kaggle/contrail_detector/data/train_metadata.json", dtype={'record_id': 'str'})
valid_metadata = pd.read_json("/content/drive/MyDrive/kaggle/contrail_detector/data/valid_metadata.json", dtype={'record_id': 'str'})
test_metadata = pd.read_json("/content/drive/MyDrive/kaggle/contrail_detector/data/test_metadata.json", dtype={'record_id': 'str'})
```

```python
import numpy as np
def contrail_exists(record_id, mask_folder_path) :
  mask = np.load(mask_folder_path + str(record_id) + ".npy")
  return int(np.sum(mask) > 0)
```

```python
i=2
mask = np.load("/content/drive/MyDrive/kaggle/contrail_detector/data/train/mask/"+ str(train_metadata.record_id.iloc[i]) + ".npy")
np.sum(mask)
```

```python
contrail_exists(train_metadata.record_id.iloc[i], "/content/drive/MyDrive/kaggle/contrail_detector/data/train/mask/")
```

```python
train_metadata['contrail_exists'] = train_metadata['record_id'].apply(lambda x: contrail_exists(x, "/content/drive/MyDrive/kaggle/contrail_detector/data/train/mask/"))
valid_metadata['contrail_exists'] = valid_metadata['record_id'].apply(lambda x: contrail_exists(x, "/content/drive/MyDrive/kaggle/contrail_detector/data/valid/mask/"))
test_metadata['contrail_exists'] = test_metadata['record_id'].apply(lambda x: contrail_exists(x, "/content/drive/MyDrive/kaggle/contrail_detector/data/test/mask/"))
```

```python
train_metadata.to_json("/content/drive/MyDrive/kaggle/contrail_detector/data/train_metadata.json")
valid_metadata.to_json("/content/drive/MyDrive/kaggle/contrail_detector/data/valid_metadata.json")
test_metadata.to_json("/content/drive/MyDrive/kaggle/contrail_detector/data/test_metadata.json")
```
